```python
import pandas as pd
import numpy as np
import requests
from census import Census
from us import states
```

```python
rename_dict = {
     'NAME': 'name',
     'B01001_001E': "population",
     'B02001_002E': "white_total", # for all races chose people who identified only as that race B01001A_001E
     'B02001_003E': "black_total",
     'B02001_004E': "native_american_total",
     'B02001_005E': "asian_total",
     'B02001_006E': "hawaiian_total",
     'B02001_007E': "some_other_race_alone_total",
     'B02001_008E': "two_more_races_total",
     'B03002_012E': "hispanic_or_latino_total", 
     'B03002_013E':	"hispanic_or_latino_white_total", #Estimate!!Total:!!Hispanic or Latino:!!White alone
     'B03002_014E': "hispanic_or_latino_black_total",#	Estimate!!Total:!!Hispanic or Latino:!!Black or African American alone
     'B03002_015E':	"hispanic_or_latino_american_indian_total",
     'B03002_016E':	"hispanic_or_latino_asian_total",
     'B03002_017E':	"hispanic_or_latino_hawaiian_total",
     'B03002_018E':	"hispanic_or_latino_some_other_race_total",
     'B03002_019E':	"hispanic_or_latino_two_or_more_races_total",
     'B03003_002E': 'not_hispanic_or_latino_total'
}

categories = list(rename_dict.keys())
```

```python
c = Census('307a9c458f89f32c9cf817fb1ae4f426ccc84e2c')
#This is the key to use the API. It was obtained at https://api.census.gov/data/key_signup.html
```

```python
timeperiod = 2019

acs_5yr = c.acs5.state_county(
    categories,
    Census.ALL,
    Census.ALL,
    year = timeperiod
    )

df_county = pd.DataFrame(acs_5yr)
df_county = df_county.rename(columns=rename_dict)

df_county.head()
```

```python
df_county.columns

df_county[['white_total', 'black_total',
       'native_american_total', 'asian_total', 'hawaiian_total',
       'some_other_race_alone_total', 'two_more_races_total',]].sum(axis=1)

df_county[['not_hispanic_or_latino_total', 'hispanic_or_latino_total']].sum(axis=1)
```
