```python

homeless_df = pd.read_csv(f'{data_path}/processed/homeless_students_201819-totals.csv')
homeless_df.drop_duplicates(inplace=True)
homeless_df.head()

homeless_df[homeless_df['FIPS']==51059] #2467.0
homeless_df[homeless_df['FIPS']==51600] #2467.0

main_df[main_df['FIPS'].isin([51059, 51600])]['public_students_pre_12'].sum()
```

```python
main_df = main_df.merge(homeless_df, how='left', left_on='FIPS', right_on='FIPS')
```

```python

# fairfax city county, virginia shares school districts with fairfax county, VA so 
# they need to be handled appropriately 
# one solution is to set the total public school students to the sum of Fairfax County and Fairfax city public school students
# then the calculations should be more reasonable
main_df.loc[main_df['FIPS'] == 51059, 'HOM_STUDENTS'] = main_df[main_df['FIPS']==51059]['HOM_STUDENTS']* main_df[main_df['FIPS'].isin([51059])]['public_students_pre_12']/main_df[main_df['FIPS'].isin([51059, 51600])]['public_students_pre_12'].sum()
```

```python
main_df.loc[main_df['FIPS'] == 51600, 'HOM_STUDENTS'] = main_df[main_df['FIPS']==51600]['HOM_STUDENTS']* main_df[main_df['FIPS'].isin([51600])]['public_students_pre_12']/main_df[main_df['FIPS'].isin([51059, 51600])]['public_students_pre_12'].sum()
```

```python

main_df['proportion_homeless'] = main_df['HOM_STUDENTS']/main_df['public_students_pre_12']
main_df.sample(5)

main_df['proportion_homeless'].sort_values(ascending=False)

main_df[['NAME', 'public_students_pre_12','HOM_STUDENTS']].loc[2475]
```
