```python
import pandas as pd
data_path = 'data'
```

```python
df = pd.read_csv(f'{data_path}/interim/crime_data_fips_split.csv')
```

```python
df
```

```python
import numpy as np
# format the df and prepare for merge

formatted_crime_df = df[['FIPS', '2019']]
formatted_crime_df = formatted_crime_df.replace('NOFIPS', np.nan)
formatted_crime_df = formatted_crime_df.rename(columns={'2019': 'crime'})
```

```python
#formatting so we can easily merge with the other dfs
formatted_crime_df['FIPS'] = formatted_crime_df['FIPS'].fillna(-1).astype(int).replace(-1, np.nan)
```

```python
formatted_crime_df
```

```python
formatted_crime_df.to_csv(f'{data_path}/processed/crime_data_fips_split.csv', index=False)
```
